```python
### Define the space
N = int(1000)
xmin,xmax = 0,1
x = np.linspace(xmin,xmax,N)
dx = x[1]-x[0]

### Define the time
tmax = 3; t = 0
dt = 0.05; 
ts = 1; count = 0

### Define b vector
b1 = 1
b2 = 0
b = np.zeros(N)

### Construct A
A2 = dt/dx**2*(np.roll(np.eye(N),-1,axis=0) + np.roll(np.eye(N),1,axis=0) - 2 * np.eye(N))

A = np.eye(N)-A2

### Apply BC
A[0,:] = 0
A[0,0] = 1
A[-1,:] = 0
A[-1,-1] = 3/2*1/dx
A[-1,-2] = -2*1/dx
A[-1,-3] = 1/2*1/dx
#print("AAAA")
#print(A)
#print("AAAA")
print(A)
### Invert A
Ainv = np.linalg.inv(A)
### Define IC
f = np.exp(-5*x)
f[0] = 1
f[-1] = 0

### Solve the system
while t < tmax:

    bnew = f.copy()
    bnew[0] = b1
    bnew[-1] = b2
    f = Ainv@bnew
    if (t>ts*count):
        plt.plot(x,f,label = f't = {t:.2f}')
        count+=1
    t+=dt
plt.plot(x,f,label = 't=3')
plt.legend()
plt.show()


```

```python
N = 1000

# Define the function and its derivative
def f(x, a):
    return np.where(x < 0, np.exp(-x) + a * x - 1,np.where(x>0, x**2,0))
   

def f_diff(x, a):
    return np.where(x < 0, -np.exp(-x) + a, np.where(x>0, 2 * x,0))

# Create an array of x values
x = np.linspace(-1, 1, N, 400)
a0=f(x,0)
a1=f(x,1)
a2=f(x,2)
a0_diff=f_diff(x,0)
a1_diff=f_diff(x,1)
a2_diff=f_diff(x,2)

# Plot for different values of a
plt.figure(figsize=(10, 5))
plt.plot(x, a0, label='a=0')
plt.plot(x, a1, label='a=1')
plt.plot(x, a2, label='a=2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function for different values of a')
plt.legend()
plt.grid(True)
plt.show()

# Plot the derivatives for different values of a
plt.figure(figsize=(10, 5))
plt.plot(x, a0_diff, label='a=0 derivative')
plt.plot(x, a1_diff, label='a=1 derivative')
plt.plot(x, a2_diff, label='a=2 derivative')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.title('Derivatives for different values of a')
plt.legend()
plt.grid(True)
plt.show()
```

```python
dx = x[1] - x[0]
a_values = [0, 1, 2]
A = np.zeros((N, N))

# Forward difference
A[0, 0] = -3 / 2
A[0, 1] = 2
A[0, 2] = -1 / 2

# Central difference
for i in range(1, N - 1):
    A[i, i - 1] = -1/2
    A[i, i + 1] = 1/2

# Backward difference
A[N - 1, N - 3] = 1 / 2
A[N - 1, N - 2] = -2
A[N - 1, N - 1] = 3 / 2

for a in a_values:
    f_analytical = f_diff(x, a)
    f_central_scheme = A @ f(x, a) / dx

    plt.plot(x, f_analytical, label='Analytical derivative, (a={})'.format(a))
    plt.plot(x[::5], f_central_scheme[::5],'x', label='Numerical derivative, (a={})'.format(a))  
plt.title('Comparison of Analytical and Numerical Derivatives')
plt.xlabel('x')
plt.ylabel("f ' (x)")
plt.legend()
plt.grid(True)
plt.xlim(-0.1, 0.1)  
plt.show()
```
